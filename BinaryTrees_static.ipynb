# Binary Trees (static version)


En una clase, un método estático está ligado a la clase y no al objeto. Un método estático no opera sobre los atributos de un objeto, sino que sólo opera sobre parámetros o variables locales del mismo método. 

En python, para invocar un método estático no es necesario crear un objeto de la clase, sino que directamenet se escribirá el nombre de la clase seguido de un punto y seguido del nombre del método estático. 

Los métodos estáticos no llevan el parámetro self, pero deben incluir la etiqueta @staticmethod antes de su declaración.
